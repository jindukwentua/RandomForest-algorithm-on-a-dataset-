# DATA CLEANING / FEATURE ENGINEERING

```python
amazon
```

```python
amazon = amazon.drop(['date', 'rating'], axis = 1)
```

```python
amazon
```

```python
new_variation = pd.get_dummies(amazon['variation'], drop_first = True)
```

```python
new_variation 
```

```python
amazon.drop(['variation'], axis = 1, inplace = True)
```

```python
amazon
```

```python
amazon = pd.concat([amazon, new_variation], axis = 1)
```

```python
amazon
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
amazon_count_vec = vectorizer.fit_transform(amazon['verified_reviews'])
```

```python
amazon_count_vec.shape
```

```python
print(vectorizer.get_feature_names())
```

```python
print(amazon_count_vec.toarray())
```

```python
amazon.drop(['verified_reviews'], axis = 1, inplace = True)
```

```python
amazon
```

```python
encoded_variation = pd.DataFrame(amazon_count_vec.toarray())
```

```python
amazon = pd.concat([amazon, encoded_variation], axis = 1)
```

```python
amazon
```

# DATA TRAINING

```python
X = amazon.drop(['feedback'], axis = 1)
```

```python
X.shape
```

```python
y = amazon['feedback']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
 from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
```

```python
randomforest_classifier = RandomForestClassifier(n_estimators = 50, criterion = 'entropy')
randomforest_classifier.fit(X_train, y_train)
```

```python
y_predict_train = randomforest_classifier.predict(X_train)
```

```python
y_predict_train
```

```python
y_train
```

```python
cm = confusion_matrix(y_train, y_predict_train)
```

```python
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_train, y_predict_train))
```

```python
y_predict = randomforest_classifier.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test, y_predict))
```

# IMPROVE THE MODEL

```python
amaz = pd.read_csv('amazon_alexa.csv', sep = '\t')
```

```python
amaz = pd.concat([amaz, pd.DataFrame(amazon_count_vec.toarray())], axis =1)
```

```python
amaz
```

```python
amaz['length'] = amaz['verified_reviews'].apply(len)
```

```python
amaz
```

```python
X = amaz.drop(['rating', 'date', 'variation', 'verified_reviews', 'feedback'], axis = 1)
```

```python
X
```

```python
y = amaz['feedback']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 101)
```

```python
randomforest_classifier = RandomForestClassifier(n_estimators = 50, criterion = 'entropy')
randomforest_classifier.fit(X_train, y_train)
```

```python
y_predict = randomforest_classifier.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test, y_predict))
```
